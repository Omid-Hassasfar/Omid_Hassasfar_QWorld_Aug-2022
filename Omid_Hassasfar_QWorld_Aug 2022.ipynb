## Simple Plot
#### Create a simple plot
https://www.machinelearningplus.com/plots/matplotlib-tutorial-complete-guide-python-plot-examples/

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 4, 0.01)
s = 1 + np.sin(2 * np.pi * t)
z = 1 + np.cos(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, s,  marker='o', markersize=2, label='Sin')
ax.plot(t, z , label='Cos')

ax.set(xlabel='time (s)', ylabel='Voltage (mV)',title='About as simple as it gets, folks',label="Line 1")
ax.grid()

ax.legend(loc='best')

plt.xlim(0, 3)
plt.ylim(0, 2.2)

fig.savefig("test.png")
plt.show()
```

```python
uv = 0; # summation is initially zero

for i in range(len(u)): # iteratively access every pair with the same indices
    uv = uv + u[i]*v[i] # i-th entries are multiplied and then added to summation
print("The dot product of",u,'and',v,'is',uv)  
```
